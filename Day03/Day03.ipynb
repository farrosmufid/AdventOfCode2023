```python
# Day 03: Gear Ratios Part 2 - Example

with open("input_example.txt") as file:
    data = file.read()
    lines = data.strip().split("\n")
```

```python
lines
```

- Idea:
    - Check every number, if found star, save the index of the star
    - If the index of the star already exist, means it overlaps with the number, so multiply with the number
    - Key = star index in string
    - Value = value of the number, save it in an array

```python
m = len(lines[0])
n = len(lines)

def is_symbol(i, j):
    # Check out of bounds
    if not(0 < i < m and 0 < j < n):
        return False
    
    return lines[i][j] != '.' and not lines[i][j].isnumeric()
```

```python
m = len(lines[0])
n = len(lines)

def is_star(i, j):
    # Check out of bounds
    if not(0 < i < m and 0 < j < n):
        return False
    
    return lines[i][j] == "*"
```

```python
star_dict = {}
```

```python
num_str = ''

num_arr = []

for i in range(0, len(lines)):
    
    start_flag = True

    for j in range(0, len(lines[0])):

        if lines[i][j].isnumeric():
            if start_flag:
                start_j = j
            num_str += lines[i][j]
            start_flag = False
            
        if not lines[i][j].isnumeric():
            end_j = j - 1
            start_flag = True

            symbol_exist_flag = False
            star_exist_flag = False

            if num_str: 
                print("num_str: {}, start_j: {}, end_j: {}".format(num_str, start_j, end_j))
                
                # Check left
                left_start_j = start_j - 1
                if is_symbol(i, left_start_j):
                    symbol_exist_flag = True

                if is_star(i, left_start_j):
                    star_exist_flag = True
                    if '{}{}'.format(i, left_start_j) not in star_dict:
                        star_dict['{}{}'.format(i, left_start_j)] = [int(num_str)]
                    else:
                        star_dict[['{}{}'.format(i, left_start_j)]].append(int(num_str))


                # Check right
                right_end_j = end_j + 1
                if is_symbol(i, right_end_j):
                    symbol_exist_flag = True

                if is_star(i, right_end_j):
                    star_exist_flag = True
                    if '{}{}'.format(i, right_end_j) not in star_dict:
                        star_dict['{}{}'.format(i, right_end_j)] = [int(num_str)]
                    else:
                        star_dict['{}{}'.format(i, right_end_j)].append(int(num_str))

                # Check top
                top_i = i - 1
                top_left_j = start_j - 1
                top_right_j = end_j + 1

                for j_top in range(top_left_j, top_right_j + 1):
                    if is_symbol(top_i, j_top):
                        symbol_exist_flag = True

                    if is_star(top_i, j_top):
                        star_exist_flag = True
                        if '{}{}'.format(top_i, j_top) not in star_dict:
                            star_dict['{}{}'.format(top_i, j_top)] = [int(num_str)]
                        else:
                            star_dict['{}{}'.format(top_i, j_top)].append(int(num_str))
                
                # Check bottom
                bottom_i = i + 1
                bottom_left_j = start_j - 1
                bottom_right_j = end_j + 1

                for j_bottom in range(bottom_left_j, bottom_right_j + 1):
                    if is_symbol(bottom_i, j_bottom):
                        symbol_exist_flag = True

                    if is_star(bottom_i, j_bottom):
                        star_exist_flag = True
                        if '{}{}'.format(bottom_i, j_bottom) not in star_dict:
                            star_dict['{}{}'.format(bottom_i, j_bottom)] = [int(num_str)]
                        else:
                            star_dict['{}{}'.format(bottom_i, j_bottom)].append(int(num_str))

                if symbol_exist_flag:
                    num_int = int(num_str)
                    num_arr.append(num_int)
                        
            num_str = ''
```

```python
star_dict
```

```python
res = 0

for key, val in star_dict.items():
    if len(val) > 1:
        mult_res = 1

        for v in val:
            mult_res *= v

        res += mult_res
```

```python
res
```
